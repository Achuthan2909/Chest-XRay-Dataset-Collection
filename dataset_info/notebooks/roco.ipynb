```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Load the datasets
train_df = pd.read_csv('/home/somehow/projects/Medical-Report-Generation/remote-dataset/roco-dataset/all_data/train/radiologytraindata.csv')
test_df = pd.read_csv('/home/somehow/projects/Medical-Report-Generation/remote-dataset/roco-dataset/all_data/test/radiologytestdata.csv')
val_df = pd.read_csv('/home/somehow/projects/Medical-Report-Generation/remote-dataset/roco-dataset/all_data/validation/radiologyvaldata.csv')

# Extract unique labels
train_labels = set(train_df['caption'])
test_labels = set(test_df['caption'])
val_labels = set(val_df['caption'])

# Venn diagram for label overlap
plt.figure(figsize=(8, 6))
venn3([train_labels, test_labels, val_labels], ('Train', 'Test', 'Validation'))
plt.title("Venn Diagram of Label Overlap")
plt.show()

```
